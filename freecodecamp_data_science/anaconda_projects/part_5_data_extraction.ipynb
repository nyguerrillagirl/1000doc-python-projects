# 3. Selecting elements by index position with iloc()

```python
# set up df from scratch
df = pd.read_csv('players_20.csv')

# set index
df.set_index('short_name', inplace=True)

# select columns
df = df[['long_name','age','dob','height_cm','weight_kg','nationality','club']]
```

## 3.1 Selecting with a single value

iloc[row_position, column_position]

```python
# get the height of L. Messi
df.iloc[0, 3]
```

```python
# get the weight of Cristiano Ronaldo
df.iloc[1, 4]
```

```python
# get all rows inside the 'height_cm' colulmn
df.iloc[:,3]
```

```python
# get all columns that correspond to the index 'L. Messi'
df.iloc[0,:]
```

## 3.2 Selecting with a list of values

```python
# get all data about 'L. Messi' and 'Cristiano Ronaldo'
df.iloc[[0,1]]
```

```python
# get the height of 'L. Messi' and 'Cristiano Ronaldo'
df.iloc[[0,1], 3]
```

```python
# get the height and weight of 'L. Messi'
df.iloc[0, [3,4]]
```

```python
# get the height and weight of 'L. Messi' and 'Cristiano Ronaldo'
df.iloc[[0,1], [3,4]]
```

## 3.3 Selecting a range of data with a slice

start:stop (Note: start in included by the stop is excluded.)

```python
# slice column labels age/club
# players = ['L. Messi' and 'Cristiano Ronaldo']
players = [0,1]
df.iloc[players, 1:7]
```

```python
# slice index labels (top1 to top10 player)
# columns ['age', 'dob', 'height_cm', 'weight_kg']
columns = [1,2,3,4]

# get top1 and top10 player name
df.iloc[0:10, columns]
```

## 3.4 Selecting with conditions

iloc cannot accept a boolean Series, but only a boolean list. We have to use the list() function to convert a Series into a boolean list.

```python
# one condition: select player with height above 180cm
# columns = ['age', 'dob', 'height_cm', 'weight_kg']
columns = [1,2,3,4]

#list(df['height_cm']>180)

df.iloc[list(df['height_cm']>180), columns]
```

```python
# multiple conditions: select player with height above 180cm and from Argentina
# list( (df['height_cm']>180) & (df['nationality']=="Argentina") )

df.iloc[list( (df['height_cm']>180) & (df['nationality']=="Argentina") ), :]
```

# 7. Filter A DataFrame with the query() method

```python
# restore the dataframe information
# reads csv file
df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name', inplace=True)
# select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
df.head()
```

## 7.1 query()

```python
# select players older than 34
df.query("age>34")
```

```python
# exercise: write the equivalent boolean slicing
df[df['age'] > 34].head()
```

```python
# select players older than 34 form Italy
df.query("age>34 and nationality=='Italy'")
```

```python
# exercise: write the equivalent boolean slicing
df[(df['age'] > 34) & (df['nationality'] == 'Italy')].head()
```

```python
# add a not operator to the first example
df.query("not (age>34)")
```

```python
# exercise: write the equivalent using boolean slicing
df[~(df['age'] > 34)].head()
```

```python
# convert height to meters and select those with height above 1.8
df.query("(height_cm / 100) > 1.8")
```

```python
# exercise: write the equivalent using boolean slicing
df[ (df["height_cm"] / 100) > 1.8]
```

```python
# select players that wer born after 1990

# check out data type
df.dtypes
# convert "dob" column to datetime type
df['dob'] = df['dob'].astype('datetime64[ns]')

# query
df.query("dob.dt.year > 1990")
```

```python
# exercise: write the equivalent using boolean slicing
df[ df["dob"].dt.year > 1990]
```
