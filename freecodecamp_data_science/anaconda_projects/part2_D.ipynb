# 1. Adding a new column

```python
from scipy.stats import beta
import numpy as np
```

```python
a, b = 5, 3 # Right-skewed, 
size = len(df_exams)
```

```python
# Generate values from beta distribution scaled to 0-100
np.random.seed(42)
raw_scores = beta.rvs(a, b, size=size) * 100
```

```python
# Shift scores so median is ~70
median_shift = 70 - np.median(raw_scores)
adjusted_scores = raw_scores + median_shift
```

```python
# clip scores to stay within 0-100
adjusted_scores = np.clip(adjusted_scores, 0, 100)
df_exams['language score'] = adjusted_scores
```

```python
df_exams.describe()
```

# 1. Adding a new column

## 1.1 assign()

When to use assign?

- Add multiple columns in a single line of code
- When you need to overwrite the values of an existing columns (best practice)

It returns a new object (copy) with all the original columns in addition to the new ones

```python
# create random numbers for our new columns
score1 = np.random.randint(1, 100, size=1000)
score2 = np.random.randint(1, 100, size=1000)
```

```python
# create a series using the random numbers
series1 = pd.Series(score1, index=np.arange(0, 1000))
series2 = pd.Series(score2, index=np.arange(0, 1000))
```

```python
df_exams.assign(score1=series1, score2=series2)
```

## 1.2 insert()

Inserts a new column at a specific position or index

```python
# using insert() to add a column at a specific position
df_exams.insert(1, 'test', series1)
```
