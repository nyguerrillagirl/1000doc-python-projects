# 2 pivot_table()

```python
# read and show the dataset
df_sales = pd.read_excel('supermarket_sales.xlsx')
df_sales.head()
```

```python
df_sales.info()
```

```python
df = pd.DataFrame({
    'date': ['2025-08-01', '2025-08-01', '2025-08-02', '2025-08-02'],
    'city': ['New York', 'Chicago', 'New York', 'Chicago'],
    'temperature': [85, 80, 88, 82]
})
df
```

```python
df.info()
```

```python
pivoted = df.pivot(index='date', columns='city', values='temperature')
pivoted
```

```python
# pivot example with muliple values (not allowed)
df = pd.DataFrame({
    'date': ['2025-08-01', '2025-08-01'],
    'city': ['New York', 'New York'],
    'temperature': [85, 86]
})
# This will raise a ValueError due to duplicate index/column pairs
#df.pivot(index='date', columns='city', values='temperature')

# Fix
df.pivot_table(index='date', columns='city', values='temperature', aggfunc='mean')
```

```python
# make a pivot table, add an aggregate function
# pivot_table(index="Gender", columns="Total", aggfunc="sum")
df_sales_sans_date = df_sales.drop(columns=['Date', 'Time'], errors='ignore')
#df_sales_sans_date.info()
df_sales_sans_date.pivot_table(index="Gender", values=["Quantity", "Rating", "Total", "Unit price", "cogs", "gross income"],aggfunc="sum")
```

```python
# make a pivot table, add an aggregate function and select some columns
df_sales_sans_date.pivot_table(index="Gender", values=["Quantity", "Total"],aggfunc="sum")
```

```python
# make a pivot table that says how much male and female spend in each 
df_sales_sans_date.pivot_table(index="Gender", columns="Product line", values=["Total"],aggfunc="sum")
```
