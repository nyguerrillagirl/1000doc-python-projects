```python
import pandas as pd

# reading the csv file
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

# 1 Math Operations

## 1.1 Operations in columns

```python
# select a column and calculate total sum
df_exams['math score'].sum()
```

```python
# count, mean, std, max, and min
```

```python
df_exams['math score'].count()
```

```python
df_exams['math score'].sum() / df_exams['math score'].count()
```

```python
df_exams['math score'].mean()
```

## 1.2 Operation in rows

```python
# calculating the sum in a row
df_exams['math score'] + df_exams['reading score'] + df_exams['writing score'] 
```

```python
# calculating the average score and assigning the result to a new column
df_exams['average score'] = df_exams['math score'] + df_exams['reading score'] + df_exams['writing score'] / 3
```

```python
# showing the dataframe
df_exams = df_exams.round(2)
df_exams
```

# Sort a Dataframe

```python
# sort by a column
df_exams.sort_values(by='math score')
```

```python
# sort descending by one column
df_exams.sort_values(by='math score', ascending=False)
```

```python
# sort descending by muliple columns
df_exams.sort_values(by=['math score', 'reading score'], ascending=False)
```

```python
# sort descending with a key function
df_exams.sort_values('race/ethnicity', ascending=True, key=lambda col:col.str.lower())
```

# 1 Create Index

```python
import numpy as np
import random 
```

```python
# creating non repetitive values for the index
new_index = np.arange(0, 1000)
```

```python
# shuffling the indexes
random.shuffle(new_index)
new_index
```

```python
# create a new column with new index
df_exams['new_index'] = new_index
df_exams
```

```python
# set new_index as the new index for the dataframe
df_exams.set_index('new_index')
```

```python
df_exams.set_index('new_index', inplace=True)
df_exams
```

```python
df_exams.sort_index()
```
