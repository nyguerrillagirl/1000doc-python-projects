```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='unicode_escape')
```

```python
df_laptops.head(3)
```

# 1 Filter dataframe based on a condition

```python
# find which rows have "Apple" in the "Company" column
len(df_laptops[df_laptops["Company"] == "Apple"])
```

```python
# verify which companies are in the "Company" column with value_counts()
df_laptops["Company"].value_counts()
```

## 1.1 Exercise 1

```python
# find which rows don't have "HP" in the "Company" column
df_laptops[df_laptops["Company"] != "HP"].head()
```

## 1.2 Exercise 2

```python
# find laptops with price over 2000 euros
df_laptops[df_laptops["Price_euros"] > 2000].sort_values(by="Price_euros", ascending=False).head()
```

# 1. Create a column based on 1 conditions: np.where()

```python
import numpy as np
```

```python
# create an array based on price tiers
df_laptops['Price_tier'] = np.where(df_laptops["Price_euros"] > 2000, 'Expensive', 'Cheap')
```

```python
df_laptops.head()
```

```python
# count values in price_tier col
df_laptops.value_counts('Price_tier')
```

## 1.1 Exercise

```python
# create an array based on screen size
np.where(df_laptops["Inches"] > 15, "Big", "Small")
```

```python
# set it to a new column
df_laptops["Screen_size"] = np.where(df_laptops["Inches"] > 15, "Big", "Small")
```

```python
# show first five rows
df_laptops.head()
```

```python
# count values in Screen_size column
df_laptops.value_counts("Screen_size")
```

# 1. Filter a dataframe based on multiple conditions

```python
(df_laptops['Company'] == 'Apple') & (df_laptops['Price_euros'] > 2000)
```

```python
df_laptops[(df_laptops['Company'] == 'Apple') & (df_laptops['Price_euros'] > 2000)]
```

## 2.2 | (or)

```python
# find Apple or Dell laptops
df_laptops[(df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'Dell')].head()
```

```python
df_laptops[(df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'Dell')].value_counts('Company')
```

```python
# find laptops from Apple or Dell that cost over 2000 euros
df_laptops[((df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'Dell')) & (df_laptops['Price_euros'] > 2000)].head()
```
