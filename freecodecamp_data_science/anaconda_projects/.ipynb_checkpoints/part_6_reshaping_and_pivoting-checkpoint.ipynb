# Part 6. Reshaping and Pivoting Dataframes

# 1 pivot

```python
import pandas as pd
```

```python
# read and show the dataset
df_gdp = pd.read_csv('gdp.csv', encoding='unicode_escape')
df_gdp.head()
```

```python
# reshape the dataframe with .pivot()
df_gdp.pivot(index="year", columns="country", values="gdppc")
```

# 2 pivot_table()

```python
# read and show the dataset
df_sales = pd.read_excel('supermarket_sales.xlsx')
df_sales.head()
```

```python
df_sales.info()
```

```python
df_sales.drop(columns=["Date"])
```

```python
# make a pivot table and add an aggregate function
df_sales.drop(columns=["Date"])
df_sales.pivot_table(index="Gender", aggfunc="sum")
```
