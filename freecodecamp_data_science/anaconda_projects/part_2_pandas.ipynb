# Creating a DataFrame from a dictionary

```python
# lists used for this example
states = ['California', 'Texas', 'Florida', 'New York']
population = [39613493, 29730311, 21944577, 19299981]
```

```python
# Storing lists within a dictionary
z = zip(states, population)
for i,v in z:
    print(i, " ", v)
```

```python
[{'state': i, 'population': v} for i,v in zip(states, population)]
```

```python
# Storing lists within a dictionary
dict_states = { 'States': states, 'Population': population}
dict_states
```

```python
# Creating the dataframe
df_population = pd.DataFrame(dict_states)
```

```python
# showing the dataframe
df_population
```

```python
my_population_state = [{"States": "California", "Population": 39613493 },
                       {"States": "Texas", "Population": 29730311 },
                       {"States": "Florida", "Population": 21944577 },
                       {"States": "New York", "Population": 19299981 }]
my_population_state
```

```python
df_population2 = pd.DataFrame(my_population_state)
df_population2
```
