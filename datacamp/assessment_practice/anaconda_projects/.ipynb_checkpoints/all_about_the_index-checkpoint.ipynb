### 🟦 Option 1: A tidy DataFrame with a normal RangeIndex

Perfect for practicing **set_index**, **reset_index**, **reindex**, **sort_index**, and **filtering**.
```
import pandas as pd

df = pd.DataFrame({
    "city": ["Paris", "Paris", "London", "London", "Berlin"],
    "year": [2020, 2021, 2020, 2021, 2020],
    "population": [2.1, 2.2, 8.9, 9.0, 3.6]
})

df
```


````python
df.index
````

````python
df.index.name
````

````python
df.index.dtype
````

````python
df.index.values
````

````python
df.index.to_list()
````

````python
# make a deep copy 
df_copy = df.copy()
````

````python
df_copy
````

````python
df2 = df_copy.reset_index()
df2
````

````python
df2.index
````

````python
df_copy = df.copy()
df2 = df_copy.reset_index(drop=True)
df2
````

````python
df2.index
````

````python
df_copy = df.copy()
df_copy
````

````python
df2 = df_copy.set_index("population")
df2
````

````python
df_copy = df.copy()
df2 = df_copy.sort_index(ascending=False)
df2
````

````python
df_copy = df.copy()
df2 = df_copy.rename(index={0:'p1', 1:'p2', 2:'p3', 3:'p4', 4:'p5'})
df2
````

````python
# more Pythonic way of doing the above
df_copy = df.copy()
df2 = df_copy.rename(index={x: f"p{x+1}" for x in df_copy.index.to_list()})
df2
````

````python
df2.loc['p1']
````

````python
df2.loc[['p2', 'p4']]
````

````python
df2.iloc[0]
````

````python
df2.iloc[[1, 3]]
````

````python
df3 = df2.reset_index()
df3
````

````python
df_paris = df3[df3['city'] == 'London'].reset_index()
df_paris
````

````python
df2
````

````python
df4 = df2.drop(index='p3')
df4
````

````python
df5 = df2[df2.index > 'p3']
df5
````
