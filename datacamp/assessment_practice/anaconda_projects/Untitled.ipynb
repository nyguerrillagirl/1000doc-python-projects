### 🟨 Problem 4 — Outlier Detection Using IQR
Given:
```
data = [5, 7, 8, 8, 9, 10, 10, 11, 50]
```
#### Task:
- Compute Q1, Q3, and IQR
- Compute the lower and upper bounds using the 1.5×IQR rule
- Identify any outliers


````python
data = [5, 7, 8, 8, 9, 10, 10, 11, 50]
````

````python
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
np_iqr = q3 - q1
np_iqr
````

````python
# use stats
stats_iqr = stats.iqr(data)
stats_iqr
````

````python
lower = q1 - 1.5 * np_iqr
upper = q3 + 1.5 * np_iqr
print(f"lower: {lower}, upper: {upper}")
````

````python
outliers = [x for x in data if x < lower or x > upper]
outliers
````

### 🟫 Problem 6 — NaN Handling

Given:
```
data = [1, 2, 3, np.nan, 4, 5]
```
#### Task:

Compute the IQR twice:
- stats.iqr(data, nan_policy='propagate')
- stats.iqr(data, nan_policy='omit')

#### Question:
What is the difference in the results, and why?



````python
data = [1, 2, 3, np.nan, 4, 5]
````

````python
iqr1 = stats.iqr(data, nan_policy='propagate')
print(f"iqr1 = {iqr1}")
````

````python
iqr2 = stats.iqr(data, nan_policy='omit')
print(f"iqr2 = {iqr2}")
````
