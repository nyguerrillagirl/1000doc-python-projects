```python
import numpy as np
import pandas as pd
```

```python
countries = ["Italy", "Italy", "Spain", "Spain"]
years = [2017, 2018, 2017, 2018]
row_index = pd.MultiIndex.from_arrays([countries, years], names=["country", "year"])
row_index
```

```python
# define the column index 
col_index = pd.MultiIndex.from_product(
    [["office supply", "Technology"], ["online", "onsite"]],
    names=[None, "shop"]
)
col_index
```

```python
data = np.array([[310, 123, 510, 340],[110, 100, 610, 120], [229, 200, 300, 240], [120, 220, 190, 210]])
data
```

```python
sales = pd.DataFrame(data, index=row_index, columns=col_index)
sales
```

```python
sales.columns.names
```

```python
sales.stack(future_stack=True)
```

```python
sales.stack(future_stack=True).sum(axis=1)
```
