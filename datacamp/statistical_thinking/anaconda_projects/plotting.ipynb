# 📈 Plotting the Normal PDF and Marking norm.pdf(-1)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Define the distribution
mu = 0
sigma = 1

# create x-values for the curve
x = np.linspace(-4, 4, 400)
y = norm.pdf(x, loc=mu, scale = sigma)
```

```python
# Compute the specific point at x=-1
x0 = -1
y0 = norm.pdf(x0, loc=mu, scale=sigma)
```

```python
# Plot the PDF
plt.plot(x, y, label="Normal PDF (μ=0, σ=1)")

# Mark the point at x = -1

plt.scatter([x0], [y0], color="red", zorder=5)
plt.text(x0, y0, f"({x0}, {y0:.3f})", color="red")

plt.title("Standard Normal PDF with point at x = -1")
plt.xlabel("x")
plt.ylabel("pdf(x)")
plt.grid(True)
plt.legend()
plt.show()
```

# 🧠 What this graph shows

- The blue curve is the standard normal PDF.
- The red dot is the value of norm.pdf(-1).
- You’re visualizing the height of the bell curve at that specific x‑value.

This is the most intuitive way to “draw a graph for norm.pdf(-1)” because a single PDF value by itself isn’t a graph — it’s just a number. The graph is the function, and the point is one sample from it.

# 1️⃣ Shade the area under the curve up to x = -1

This shows the density curve and the area under it from -\infty  to -1.

(Technically, the shaded area corresponds to norm.cdf(-1).)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4, 4, 400)
y = norm.pdf(x)

x0 = -1

plt.plot(x, y, label="Normal PDF")

# Shade area under curve up to x = -1
x_fill = np.linspace(-4, x0, 200)
plt.fill_between(x_fill, norm.pdf(x_fill), color="skyblue", alpha=0.5)

plt.title("Area under the curve up to x = -1")
plt.xlabel("x")
plt.ylabel("pdf(x)")
plt.grid(True)
plt.legend()
plt.show()
```

# 2️⃣ Graph with guide lines to the point (–1, pdf(–1))

This highlights the height of the curve at x=-1 using vertical and horizontal lines.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4, 4, 400)
y = norm.pdf(x)

x0 = -1
y0 = norm.pdf(x0)

plt.plot(x, y, label="Normal PDF")

# Mark the point
plt.scatter([x0], [y0], color="red", zorder=5)

# Guide lines
plt.axvline(x0, color="gray", linestyle="--")
plt.axhline(y0, color="gray", linestyle="--")

plt.text(x0, y0, f"  ({x0}, {y0:.3f})", color="red")

plt.title("Guide lines to the point x = -1")
plt.xlabel("x")
plt.ylabel("pdf(x)")
plt.grid(True)
plt.legend()
plt.show()
```

# 3️⃣ Graph of norm.cdf(-1) with only guidelines

Here we are not shading the area - just showing the value of the CDF at x=-1.
This is a horizontal line at the CDF value and a vertical line at x=-1.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Standard normal parameters
mu = 0
sigma = 1

# Create x-values for the CDF curve
x = np.linspace(-4, 4, 400)
y = norm.cdf(x, loc=mu, scale=sigma)

# Point to highlight
x0 = -1
y0 = norm.cdf(x0, loc=mu, scale=sigma)

# Plot the CDF
plt.plot(x, y, label="Normal CDF (μ=0, σ=1)")

# Mark the point at x = -1
plt.scatter([x0], [y0], color="blue", zorder=5)
plt.text(x0, y0, f"  ({x0}, {y0:.3f})", color="blue")

plt.title("Standard Normal CDF with point at x = -1")
plt.xlabel("x")
plt.ylabel("cdf(x)")
plt.grid(True)
plt.legend()
plt.show()
```

# Restaurant spending example

Let's go back to the example of the restaurant chain that has been collecting data about customer spending. Recall that the data shows that the spending is approximately normally distributed, with a mean of 3.15 and a standard deviation of 1.5 per customer, as pictured in the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 3.15
sigma = 1.5

# Range of spending values
x = np.linspace(0, 7, 400)

# PDF values
y = norm.pdf(x, loc=mu, scale=sigma)

# Plot
plt.plot(x, y, label=f"Normal PDF (μ={mu}, σ={sigma})", color="blue")

plt.title("mean = 3.15, stddev = 1.5")
plt.xlabel("x")
plt.ylabel("Probability density of x")
plt.grid(True)
plt.axvline(mu, color="gray", linestyle="--")
plt.legend()
plt.show()
```

## Adults' heights example

The heights of adults aged between 18 and 35 years are normally distributed. For males, the mean height is 70 inches with a standard deviation of 4. Adult females have a mean height of 65 inches with a standard deviation of 3.5. You can see how the heights are distributed in this plot:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Female parameters
mu_f = 65
sd_f = 3.5

# Male parameters
mu_m = 70
sd_m = 4
```

```python
# X-range that covers both distributions well
x = np.linspace(50, 85, 400)

# PDFs
y_f = norm.pdf(x, loc=mu_f, scale=sd_f)
y_m = norm.pdf(x, loc=mu_m, scale=sd_m)
```

```python
# Plot
plt.plot(x, y_f, color="red", label="Female Heights (μ=65, σ=3.5)")
plt.plot(x, y_m, color="blue", label="Male Heights (μ=70, σ=4)")

plt.title("Height Distributions: Females vs. Males")
plt.xlabel("Height (inches)")
plt.ylabel("pdf(x)")
plt.grid(True)
plt.legend()
plt.show()
```
