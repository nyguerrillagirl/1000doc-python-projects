## Data summarization

```python
books[["genre", "rating", "year"]].groupby("genre").mean()
```

```python
books[["rating", "year"]].agg(["mean", "std"])
```

```python
books.agg({"rating": ["mean", "std"], "year":["median"]})
```

```python
books.groupby("genre"). agg(
    mean_rating=("rating", "mean"),
    std_rating=("rating", "std"),
    median_year=("year", "median")
)
```

```python
sns.barplot(data=books, x="genre", y="rating")
plt.show()
```

```python
sns.barplot(data=books, x="genre", y="rating", hue="genre")
plt.show()
```

```python
unemployment[["2019", "2020"]].agg(["mean", "std"])
```

```python
# print the mean and std grouped by continent for 2019 and 2020
print(unemployment[["continent", "2019", "2020"]].groupby("continent").agg(["mean", "std"]))
```

```python
sns.barplot(data=unemployment, x="continent", y="2021", hue="continent")
plt.show()
```

```python
salaries = pd.read_csv("ds_salaries_clean.csv")
salaries.head()
```

```python
salaries.isna().sum()
```

```python
salaries.columns
```

```python
salaries["Company_Size"].value_counts()
```

```python
salaries.groupby("Company_Size")["Salary_USD"].mean()
```

```python
salaries["std_dev"] = salaries.groupby("Experience")["Salary_USD"].transform(lambda x: x.std())
```

```python
salaries[["Experience", "std_dev"]].value_counts()
```

```python
salaries["median_by_comp_size"] = salaries.groupby("Company_Size") \
    ["Salary_USD"].transform(lambda x: x.median())
```

```python
salaries[["Company_Size", "median_by_comp_size"]].head()
```
