# Sampling in Python

Sampling in Python is the cornerstone of inference statistics and hypothesis testing. It's a powerful skill used in survey analysis and experimental design to draw conclusions without surveying an entire population. In this Sampling in Python course, you’ll discover when to use sampling and how to perform common types of sampling—from simple random sampling to more complex methods like stratified and cluster sampling. Using real-world datasets, including coffee ratings, Spotify songs, and employee attrition, you’ll learn to estimate population statistics and quantify uncertainty in your estimates by generating sampling distributions and bootstrap distributions.

### Are findings from the sample generalizable?

You just saw how convenience sampling—collecting data using the easiest method—can result in samples that aren't representative of the population. Equivalently, this means findings from the sample are not generalizable to the population. Visualizing the distributions of the population and the sample can help determine whether or not the sample is representative of the population.
The Spotify dataset contains an acousticness column, which is a confidence measure from zero to one of whether the track was made with instruments that aren't plugged in. You'll compare the acousticness distribution of the total population of songs with a sample of those songs.

## Sampling Methods

It’s time to get hands-on and perform the four random sampling methods in Python: 
- simple,
- systematic,
- stratified, and
- cluster.

### Simple random and systematic sampling

**Simple random sampling** works like a raffle or lottery. We start with our population of raffle tickets or lottery balls and randomly pick them out one at a time.


Another sampling method is known as **systematic sampling**. This samples the population at regular intervals. 

```python
coffee_ratings.sample(n=5, random_state=19000113)
```

```python
sample_size = 5
pop_size = len(coffee_ratings)
print(pop_size)
```

```python
interval = pop_size // sample_size
print(interval)
```

```python
# systematic sampling - selecting rows
coffee_ratings.iloc[::interval]
```

```python
# plotting aftertaste shows there is a bias in the data listing
coffee_ratings_with_id = coffee_ratings.reset_index()
coffee_ratings_with_id.plot(x="index", y="aftertaste", kind="scatter")
plt.show()
```

```python
# re-drawing the plot after shuffling the rows
shuffled = coffee_ratings.sample(frac=1)
shuffled = shuffled.reset_index(drop=True).reset_index()
shuffled.plot(x="index", y="aftertaste", kind="scatter")
plt.show()
```

```python
# Sample 70 rows using simple random sampling and set the seed
attrition_samp = attrition_pop.sample(n=70, random_state=18900217)

# Print the sample
print(attrition_samp)
```

### Is systematic sampling OK?

Systematic sampling has a problem: if the data has been sorted, or there is some sort of pattern or meaning behind the row order, then the resulting sample may not be representative of the whole population. The problem can be solved by shuffling the rows, but then systematic sampling is equivalent to simple random sampling.


```python
# Add an index column to attrition_pop
attrition_pop_id = attrition_pop.reset_index()

# Plot YearsAtCompany vs. index for attrition_pop_id
attrition_pop_id.plot(x="index", y="YearsAtCompany", kind="scatter")
plt.show()
```

```python
# Shuffle the rows of attrition_pop
attrition_shuffled = attrition_pop.sample(frac=1)

# Reset the row indexes and create an index column
attrition_shuffled = attrition_shuffled.reset_index(drop=True).reset_index()

# Plot YearsAtCompany vs. index for attrition_shuffled
attrition_shuffled.plot(x="index", y="YearsAtCompany", kind="scatter")
plt.show()
```

## Stratified and weighted random sampling

**Stratified sampling** is a technique that allows us to sample a population that contains subgroups.

```python
top_counts = coffee_ratings['country_of_origin'].value_counts()
top_counts.head(6)
```

```python
# filter to the top 6
top_counted_countries = ["Mexico", "Colombia", "Guatemala", "Brazil", "Taiwan", "United States (Hawaii)"]

# create new dataframe with only the top 6 countries
top_counted_subset = coffee_ratings['country_of_origin'].isin(top_counted_countries)

coffee_ratings_top = coffee_ratings[top_counted_subset]
```

```python
# let's take 10% simple random sample of the dataset coffee_ratings_top
coffee_ratings_samp = coffee_ratings_top.sample(frac=0.1, random_state=2021)
coffee_ratings_samp['country_of_origin'].value_counts(normalize=True)
```

```python
# what is the distribution in the original dataframe for top 6
coffee_ratings_top['country_of_origin'].value_counts(normalize=True)
```

If we care about the proportions of each country in the sample closely matching those in the population, then we can group the data by country before taking the simple random sample. 

```python
coffee_ratings_strat = coffee_ratings_top.groupby("country_of_origin").sample(frac=0.1, random_state=2021)
coffee_ratings_strat['country_of_origin'].value_counts(normalize=True)
```

```python
# variation of stratified sample is to sample equal counts from each group
coffee_ratings_eq = coffee_ratings_top.groupby("country_of_origin").sample(n=15, random_state=2021)
coffee_ratings_eq['country_of_origin'].value_counts(normalize=True)
```

```python
# Another variation is weighted random sampling, we create a column of weights 
import numpy as np
coffee_ratings_weight = coffee_ratings_top
condition = coffee_ratings_weight['country_of_origin'] == "Taiwan"

coffee_ratings_weight['weight'] = np.where(condition, 2, 1)
coffee_ratings_weight = coffee_ratings_weight.sample(frac=0.1, weights="weight")
coffee_ratings_weight['country_of_origin'].value_counts(normalize=True)
```

```python
# Proportion of employees by Education level
education_counts_pop = attrition_pop['Education'].value_counts(normalize=True)

# Print education_counts_pop
print(education_counts_pop)
```

```python
# Proportional stratified sampling for 40% of each Education group
attrition_strat = attrition_pop.groupby('Education', observed=False).sample(frac=.4, random_state=2022)

# Print the sample
print(attrition_strat)
```

```python
# Calculate the Education level proportions from attrition_strat
education_counts_strat = attrition_strat['Education'].value_counts(normalize=True)

# Print education_counts_strat
print(education_counts_strat)
```

```python
# Get 30 employees from each Education group
attrition_eq = attrition_pop.groupby('Education', observed=False).sample(n=30, random_state=2022)

# Print the sample
print(attrition_eq)
```

```python
# Get the proportions from attrition_eq
education_counts_eq = attrition_eq['Education'].value_counts(normalize=True)

# Print the results
print(education_counts_eq)
```

## Cluster Sampling

The **stratified sampling** approach was to split the population into subgroups, then use simple random sampling on each of them. **Cluster sampling** means that we limit the number of subgroups in the analysis by picking a few of them with simple random sampling. We then perform simple random sampling on each subgroup as before.

```python
coffee_ratings['variety'] = coffee_ratings['variety'].astype('category')
```

```python
# to make this work I am removing from coffee_ratings all 'variety' values that have less than 5 rows
# otherwise I cannot sample with n=5 (not enough rows in the 'variety' population
filtered_coffee_ratings = coffee_ratings[coffee_ratings['variety'].map(coffee_ratings['variety'].value_counts()) >= 5]
```

```python
filtered_coffee_ratings.dtypes
```

```python
varieties_pop = list(filtered_coffee_ratings['variety'].unique())
len(varieties_pop)
```

```python
import random
varieties_samp = random.sample(varieties_pop, k=3)
varieties_samp
```

```python
# sampling each group
variety_condition = filtered_coffee_ratings['variety'].isin(varieties_samp)
coffee_ratings_cluster = filtered_coffee_ratings[variety_condition]
```

```python
coffee_ratings_cluster.head()
```

```python
coffee_ratings_cluster.dtypes
```

```python
coffee_ratings_cluster.loc[:,'variety'] = coffee_ratings_cluster['variety'].cat.remove_unused_categories()
```

```python
len(coffee_ratings_cluster)
```

```python
coffee_ratings_cluster['variety'].value_counts()
```

```python
coffee_ratings_cluster.groupby("variety").sample(n=5, random_state=2021)
```

### Perform cluster sampling

Now that you know when to use cluster sampling, it's time to put it into action. In this exercise, you'll explore the JobRole column of the attrition dataset. You can think of each job role as a subgroup of the whole population of employees.

```python
# Create a list of unique JobRole values
job_roles_pop = list(attrition_pop['JobRole'].unique())

# Randomly sample four JobRole values
job_roles_samp = random.sample(job_roles_pop, k=4)

# Print the result
print(job_roles_samp)
```

```python
# Filter for rows where JobRole is in job_roles_samp
jobrole_condition = attrition_pop['JobRole'].isin(job_roles_samp)
attrition_filtered = attrition_pop[jobrole_condition]
attrition_filtered.head()
```

```python
# Remove categories with no rows
attrition_filtered.loc[:,'JobRole'] = attrition_filtered['JobRole'].cat.remove_unused_categories()

# Randomly sample 10 employees from each sampled job role
attrition_clust = attrition_filtered.groupby('JobRole', observed=False).sample(n=10, random_state=2022)

# Print the sample
attrition_clust.head()
```

## Compare sampling methods

```python
top_counted_countries = ["Mexico", "Colombia", "Guatemala", "Brazil", "Taiwan", "United States (Hawaii)"]
subset_condition = coffee_ratings['country_of_origin'].isin(top_counted_countries)
```

```python
coffee_ratings_top = coffee_ratings[subset_condition]
coffee_ratings_top.shape
```

```python
coffee_ratings_srs = coffee_ratings_top.sample(frac=1/3, random_state=2021)
coffee_ratings_srs.shape
```

```python
coffee_ratings_strat = coffee_ratings_top.groupby('country_of_origin').sample(frac=1/3, random_state=2021)
coffee_ratings_strat.shape
```

In the cluster sample, we've used two out of six countries to roughly mimic frac equals one-third from the other sample types. Setting n equal to one-sixth of the total number of rows gives roughly equal sample sizes in each of the two subgroups. Using .shape again, we see that this cluster sample has close to the same number of rows: two-hundred-ninety-two compared to two-hundred-ninety-three for the other sample types.

```python
import random
top_countries_samp = random.sample(top_counted_countries, k=2)
top_countries_samp
```

```python
top_condition = coffee_ratings_top['country_of_origin'].isin(top_countries_samp)
```

```python
coffee_ratings_cluster = coffee_ratings_top[top_condition]
coffee_ratings_cluster.dtypes
```

```python
coffee_ratings_cluster.loc[:,'country_of_origin'] = coffee_ratings['country_of_origin'].astype('category')
```

```python
coffee_ratings_cluster.dtypes
```

```python
coffee_ratings_cluster['country_of_origin'].value_counts()
```

```python
coffee_ratings_cluster.loc[:,'country_of_origin'] = coffee_ratings_cluster['country_of_origin'].cat.remove_unused_categories()
```

```python
coffee_ratings_clust = coffee_ratings_cluster.groupby("country_of_origin", observed=False).sample(n=132)
coffee_ratings_clust.shape                                                                                
```

```python
# population mean
coffee_ratings_top['total_cup_points'].mean()
```

```python
# mean of sample
coffee_ratings_srs['total_cup_points'].mean()
```

```python
# mean on stratified sample
coffee_ratings_strat['total_cup_points'].mean()
```

```python
# mean on cluster sample
coffee_ratings_clust['total_cup_points'].mean()
```
