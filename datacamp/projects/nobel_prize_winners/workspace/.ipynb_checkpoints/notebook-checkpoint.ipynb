The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np
```

### Question: What is the most commonly awarded gender and birth country?

Store your answers as string variables **top_gender** and **top_country**.

```python
df = pd.read_csv("nobel.csv")
df.head()
```

```python
df['sex'].value_counts().sort_values(ascending=False)
```

```python
top_gender = df['sex'].value_counts().sort_values(ascending=False).index[0]
top_gender
```

```python
top_country = df['birth_country'].value_counts().sort_values(ascending=False).index[0]
top_country
```

### Question: Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?

Store this as an integer called **max_decade_usa**.

```python
df['decade'] = (df['year'] // 10) * 10
df.head()
```

```python
df.tail()
```

```python
count_USA = df[df['birth_country']=='United States of America'].groupby('decade').size()
```

```python
print(count_USA)
```

```python
total_count = df.groupby('decade').size()
total_count
```

```python
series_usa_percentage = (count_USA / total_count * 100).rename('percent_US_born').sort_values()
max_decade_usa = series_usa_percentage .index[-1]
max_decade_usa
```

### Question: Which decade and Nobel Prize category combination had the highest proportion of female laureates?

Store this as a dictionary called **max_female_dict** where the **decade** is the key and the **category** is the value. There should only be one key:value pair.

```python
# step 1: extract all female nobel winners
female_noble_winners = df[df['sex'] == 'Female']
female_noble_winners.head()
```

```python
# step 2: Group female winners by decade and category
female_counts = female_noble_winners.groupby(['decade', 'category']).size().rename('female_count')
```

```python
# ste 3: Find the (decade, category) with the highest count
max_pair = female_counts.idxmax()
max_value = female_counts.max()
print('max_pair: ' , max_pair)
print('max_value: ', max_value)
```

```python
max_female_dict = {}
max_female_dict[max_pair[0]] = max_value
max_female_dict
```

### Question: Who was the first woman to receive a Nobel Prize, and in what category?

Save your string answers as **first_woman_name** and **first_woman_category**.

```python
first_woman_name = female_noble_winners.sort_values(by="year").iloc[0,7]
first_woman_category = female_noble_winners.sort_values(by="year").iloc[0,0]
print('first_woman_name: ', first_woman_name)
print('first_woman_category: ', first_woman_category)
```

### Question: Which individuals or organizations have won more than one Nobel Prize throughout the years?

Store the full names in a list named **repeat_list**.

```python
# Count occurrences of each name
name_counts = df['full_name'].value_counts()
repeat_winners = name_counts[name_counts > 1]
repeat_winners
```

```python
# Count organizations that appear more than one time
org_counts = df['organization_name'].value_counts()
repeat_org_winners = org_counts[org_counts > 1]
repeat_org_winners
```

```python
repeat_list = repeat_winners.index
```
