Welcome to New York City, one of the most-visited cities in the world. There are many Airbnb listings in New York City to meet the high demand for temporary lodging for travelers, which can be anywhere between a few nights to many months. In this project, we will take a closer look at the New York Airbnb market by combining data from multiple file types like `.csv`, `.tsv`, and `.xlsx`.

Recall that **CSV**, **TSV**, and **Excel** files are three common formats for storing data. 
Three files containing data on 2019 Airbnb listings are available to you:

**data/airbnb_price.csv**
This is a CSV file containing data on Airbnb listing prices and locations.
- **`listing_id`**: unique identifier of listing
- **`price`**: nightly listing price in USD
- **`nbhood_full`**: name of borough and neighborhood where listing is located

**data/airbnb_room_type.xlsx**
This is an Excel file containing data on Airbnb listing descriptions and room types.
- **`listing_id`**: unique identifier of listing
- **`description`**: listing description
- **`room_type`**: Airbnb has three types of rooms: shared rooms, private rooms, and entire homes/apartments

**data/airbnb_last_review.tsv**
This is a TSV file containing data on Airbnb host names and review dates.
- **`listing_id`**: unique identifier of listing
- **`host_name`**: name of listing host
- **`last_review`**: date when the listing was last reviewed

```python
# Import necessary packages
import pandas as pd
import numpy as np
```

```python
# Read in *.csv file
airbnb_price = pd.read_csv('./data/airbnb_price.csv')
airbnb_price.head()
```

```python
# Read in *.xlsx file
airbnb_room_type = pd.read_excel('./data/airbnb_room_type.xlsx')
airbnb_room_type.head()
```

```python
# Read in *.tsv file
airbnb_last_review = pd.read_csv('./data/airbnb_last_review.tsv', sep='\t')
airbnb_last_review.head()
```

## Task: What are the dates of the earliest and most recent reviews? Store these values as two separate variables with your preferred names.

```python
# Let's see the format used in last_review
airbnb_last_review.info()
```

```python
# The last_review field is a string. Let's covert to date
airbnb_last_review['last_review'] = pd.to_datetime(airbnb_last_review['last_review'])
airbnb_last_review.info()
```

```python
earliest_review_date = min(airbnb_last_review['last_review'])
earliest_review_date
```

```python
lastest_review_date = max(airbnb_last_review['last_review'])
lastest_review_date
```

## Task: How many of the listings are private rooms? Save this into any variable.

```python
# Let's see how many unique values are in the dataframe airbnb_room_type['room_type']
airbnb_room_type['room_type'].value_counts()
```

```python
# Let's remap these values down. First let's convert everything to lower case
airbnb_room_type['room_type'] = airbnb_room_type['room_type'].str.lower()
airbnb_room_type['room_type'].value_counts()
```

```python
# extract 'private room' value
private_room_counts = airbnb_room_type['room_type'].value_counts()
number_of_private_rooms = private_room_counts['private room']
print("number of private rooms:", number_of_private_rooms)
```

## Task: What is the average listing price? Round to the nearest two decimal places and save into a variable.

```python
# Check the data types in the df: airbnb_price
airbnb_price.info()
```

```python
airbnb_price['price'] = airbnb_price['price'].str.strip(' dollars')
airbnb_price['price'].head()
```

```python
# Now convert to numeric
airbnb_price['price'] = airbnb_price['price'].astype(int)
airbnb_price['price'].head()
```

```python
average_listing_price = round(airbnb_price['price'].mean(), 2)
average_listing_price
```

## Task: Combine the new variables into one DataFrame called review_dates with four columns in the following order: first_reviewed, last_reviewed, nb_private_rooms, and avg_price. The DataFrame should only contain one row of values.

```python
review_info = {'first_reviewed': earliest_review_date, 'last_reviewed': lastest_review_date, 'nb_private_rooms': number_of_private_rooms, 'avg_price':average_listing_price}
review_dates = pd.DataFrame([review_info])
review_dates
```
