```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import binom
```

```python
x = binom.rvs(n=1, p=0.5, size=1000, random_state=7)
```

```python
running = np.cumsum(x) / np.arange(1, len(x) + 1)
```

```python
plt.plot(running)
plt.axhline(0.5, color='red')
plt.show()
```
