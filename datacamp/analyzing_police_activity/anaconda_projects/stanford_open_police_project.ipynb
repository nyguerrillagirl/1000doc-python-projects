```python
import numpy as pd
import pandas as pd
```

```python
ri = pd.read_csv('police.csv')
ri.head(3)
```

```python
ri.isnull().sum()
```

```python
ri.shape
```

```python
ri.drop('county_name', axis='columns', inplace=True)
```

```python
ri.head()
```

```python
ri.dropna(subset=['stop_date', 'stop_time'], inplace=True)
```

```python
# dropping rows with missing driver_gender information
ri.dropna(subset=['driver_gender'], inplace=True)
```

```python
ri.isnull().sum()
```

```python
ri.dtypes
```

```python
ri['is_arrested'].head()
```

```python
# change the dtype of is_arrested field
ri['is_arrested'] = ri.is_arrested.astype('bool')
```

```python
ri.is_arrested.dtype
```

```python
# create a new column, combining stop_date and stop_time
combined = ri.stop_date.str.cat(ri.stop_time, sep=' ')
ri['stop_datetime'] = pd.to_datetime(combined)
ri.dtypes
```

```python
ri.set_index('stop_datetime', inplace=True)
ri.index
```
