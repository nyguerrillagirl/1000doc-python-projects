- Use the .strip() method to strip duration of "minutes" and store it in the duration_trim column.
- Convert duration_trim to int and store it in the duration_time column.
- Write an assert statement that checks if duration_time's data type is now an int.
- Print the average ride duration.


```python
# Strip duration of minutes
ride_sharing_new['duration_trim'] = ride_sharing_new['duration'].str.strip("minutes")

# Convert duration to integer
ride_sharing_new['duration_time'] = ride_sharing_new['duration_trim'].astype('int')

# Write an assert statement making sure of conversion
assert ride_sharing_new['duration_time'].dtype == 'int'

# Print formed columns and calculate average ride duration 
print(ride_sharing_new[['duration','duration_trim','duration_time']])
print(ride_sharing_new['duration_time'].mean())
```

```python
# An example from co-pilot
import pandas as pd

# Create a small DataFrame with duplicate rows
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob'],
    'Age': [25, 30, 25, 35, 30],
    'City': ['NYC', 'LA', 'NYC', 'Chicago', 'LA']
})

print(df)
```

```python
df_dup = df.drop_duplicates()
df_dup
```

```python
# Create statistics dictionary for aggregation function
statistics = {'Age': 'min'}
df_unique = df_dup.groupby('Name').agg(statistics).reset_index()
df_unique
```

```python
# Find suplicated values again
duplicates = df.duplicated(subset = ['Name', 'Age', 'City'], keep=False)
duplicates
    
```

```python
duplicated_df = df_unique[duplicates == True]
duplicated_df 
```
