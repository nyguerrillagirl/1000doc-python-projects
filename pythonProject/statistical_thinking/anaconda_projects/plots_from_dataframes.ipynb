## 1. Line Plot

```python
import pandas as pd

df = pd.DataFrame({
    'Year': [2020, 2021, 2022, 2023],
    'Revenue': [100, 150, 200, 250]
})
df.plot(x='Year', y='Revenue', kind='line', title='Revenue Over Time')
```

```python
stock_df = pd.read_csv("appl2.csv")
stock_df.head()
```

```python
stock_df.info()
```

```python
stock_df['Date'] = pd.to_datetime(stock_df['Date'])
```

```python
# Plot closing price over time
import matplotlib.pyplot as plt
%matplotlib inline
stock_df.plot(x='Date', y='Close', kind='line', title='Apple Stock Prices Over Time', figsize=(10,5))
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.grid(True)
plt.show()
```

## 2. Bar plots

```python
df = pd.DataFrame({'Fruit': ['Apple', 'Banana', 'Cherry'], 'Count': [10, 15, 7]})
df.set_index('Fruit').plot(kind='bar')
```

```python
import pandas as pd

# Sample real-world-style data
data = {
    'Region': ['North', 'South', 'East', 'West', 'Central'],
    'Sales': [125000, 98000, 143000, 112000, 101500]
}

df = pd.DataFrame(data)

# Bar plot
df.set_index('Region').plot(kind='bar', title='Sales by Region', legend=False)
```
