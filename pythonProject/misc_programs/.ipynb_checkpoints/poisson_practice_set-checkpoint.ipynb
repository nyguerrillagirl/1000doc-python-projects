A help desk receives an average of 4 calls per hour.

```python
poisson.pmf(k=3, mu=4)
```

Answer: x=3 is more likely

# 🟧 3. SF (Survival Function = Probability of More Than k)

### Problem 5 - Machine Defects

A machine produces 0.2 defects per hour.

1. Compute P(X>1).
2. Explain why this probability is small or large based on the mean.


```python
poisson.sf(k=1, mu=0.2)
```

```python
# let's try to visualize how the probability changes for mu ranges from [0.1, ..1.0]
# Creating plot of P(X > 1) for Poisson distribution as mean varies from 0.1 to 1.0
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define range of means from 0.1 to 1.0
mu_values = np.arange(0.1, 1.1, 0.1)
# Compute P(X > 1) = sf(1, mu) for each mean
probabilities = [poisson.sf(1, mu) for mu in mu_values]

# Plotting
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(8, 5))
plt.plot(mu_values, probabilities, marker='o', color='blue', linewidth=2)
plt.title('P(X > 1) for Poisson Distribution as Mean Varies')
plt.xlabel('Mean (μ)')
plt.ylabel('P(X > 1)')
plt.grid(True)
plt.tight_layout()
plt.savefig('poisson_sf_plot.png')

print("Generated plot showing how P(X > 1) changes with Poisson mean from 0.1 to 1.0 and saved as poisson_sf_plot.png")
```

Answer: As the mean increases the P(X>1) increases.

🎯 4. Why this graph is so useful

DataCamp often teaches Poisson as:
- “rare events”
- “counts per interval”
- “use SciPy functions, don’t worry about formulas”

But the shape of this curve gives you intuition:
- When the mean is tiny (0.1), the chance of more than 1 event is almost zero.
- As the mean approaches 1, the chance of more than 1 event becomes significant.
- This helps you understand how Poisson behaves without memorizing formulas.


### Problem 6 — Server Traffic

A server receives 15 requests per second.

1. Compute P(X>20).
2. What does this tell you about the tail of the distribution


```python
poisson.sf(k=20, mu=15)
```

```python
k_values = np.arange(1, 26, 1)
sf_values = [poisson.sf(k=x, mu=15) for x in k_values]
# Plotting
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(8, 5))
plt.plot(k_values, sf_values, marker='o', color='blue', linewidth=2)
plt.title('P(X > K) for Poisson Distribution as K Varies')
plt.xlabel('K')
plt.ylabel('μ = 15')
plt.grid(True)
plt.tight_layout()
plt.savefig('poisson_sf_plot2.png')

print("Generated plot showing how P(X > K) changes with Poisson mean set at 15 and saved as poisson_sf_plot2.png")
```

2. As K increases the probabily decreases towards 0 at the tail

# 🟪 5. Mixed Practice (pmf + cdf + sf + ppf)

### Problem 9 — Hospital ER Arrivals

A hospital ER sees 5 arrivals per hour.

1. Compute P(X=4).
2. Compute P(X <= 4).
3. Compute P(X>7).
4. Find the 99th percentile of the distribution


```python
# 1. Compute P(X=4)
poisson.pmf(k=4, mu=5)
```

```python
# 2. Computer P(X <= 4)
poisson.cdf(k=4, mu=5)
```

```python
# 3. Compute P(X>7)
poisson.sf(k=7, mu=5)
```

```python
# 4. Find the 99th percentile of the distribution
poisson.ppf(q=0.99, mu=5)
```

### Problem 10 — Email Spam Filter

A user receives 3 spam emails per hour.

1. Compute P(X=0).
2. Compute P(X <= 2).
3. Compute P(X>5).
4. Compute the 95th percentile.
5. Based on your results, describe the skewness of the distribution.


```python
# 1. Compute P(X=0)
poisson.pmf(k=0, mu=3)
```

```python
# 2. Compute P(X <= 2)
poisson.cdf(k=2, mu=3)
```

```python
# 3. Compute P(X>5).
poisson.sf(k=5, mu=3)
```

```python
# 4. Compute the 95th percentile
poisson.ppf(q=0.95, mu=3)
```

- Skewness description:
The Poisson with mean 3 is right‑skewed:
- Most mass is around 2–4.
- There is a long, thin tail to the right (larger counts are possible but increasingly rare
