```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'category' : ['A', 'A', 'B', 'B'],
    'sales' : [10, 20, 5, 15],
    'quantity' : [1, 2, 1, 3]
})
df
```

```python
grouped = df.groupby('category').agg({
    'sales': ['sum', 'mean'],
    'quantity': ['min', 'max']
})
grouped
```

```python
for col in grouped.columns:
    print(col)
```

```python
grouped.columns = ['_'.join(col) for col in grouped.columns] 
grouped.columns
```
