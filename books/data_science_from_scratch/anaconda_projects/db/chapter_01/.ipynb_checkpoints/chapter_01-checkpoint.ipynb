```python
users = [
    { "id": 0, "name": "Hero" },
    { "id": 1, "name": "Dunn" },
    { "id": 2, "name": "Sue" },
    { "id": 3, "name": "Chi" },
    { "id": 4, "name": "Thor" },
    { "id": 5, "name": "Clive" },
    { "id": 6, "name": "Hicks" },
    { "id": 7, "name": "Devin" },
    { "id": 8, "name": "Kate" },
    { "id": 9, "name": "Klein" }
]
```

```python
users[0]
```

```python
users[1]
```

```python
friendship_pairs = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4),
                    (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)]
```

```python
friendships = {user["id"]:[] for user in users}
```

```python
print(friendships)
```

```python
# loop over the friendship pairs to populate it:
for i, j in friendship_pairs:
    friendships[i].append(j)
    friendships[j].append(i)
```

```python
print(friendships)
```

```python
def number_of_friends(user):
    """
    How many friends does _user_ have?
    """
    user_id = user["id"]
    friends_ids = friendships[user_id]
    return len(friends_ids)
```

```python
number_of_friends(users[0])
```

```python
number_of_friends(users[8])
```

```python
total_connections = sum([number_of_friends(user) for user in users])
print(total_connections)
```

```python
num_users = len(users)
avg_connections = total_connections / num_users
print(avg_connections)
```

```python
# find the most connected people by sorting from "most friends" to "least friends"
num_friends_by_id = [(user["id"], number_of_friends(user)) for user in users]
print(num_friends_by_id)
```

```python
num_friends_by_id.sort(key=lambda id_and_friends: id_and_friends[1], reverse=True)
print(num_friends_by_id)
```

```python
# find friends of friends
def foaf_ids_bad(user):
    """foaf is short for "friend of a friend" """
    return [foaf_id
            for friend_id in friendships[user["id"]]
            for foaf_id in friendships[friend_id]]
```

```python
foaf_ids_bad(users[0])
```

```python
from collections import Counter
```

```python
def friends_of_friends(user):
    user_id = user["id"]
    return Counter(
        foaf_id
        for friend_id in friendships[user_id]     # For each of my friends,
        for foaf_id in friendships[friend_id]     # find their friends
        if foaf_id != user_id                     # who aren't me
        and foaf_id not in friendships[user_id]   # and aren't my friends.
    )
```

```python
friends_of_friends(users[0])
```

```python
friends_of_friends(users[3])
```

```python
def data_scientists_who_like(target_interest):
    """
    Find the ids of all users who like the target interest
    """
    return [user_id for user_id, user_interest in interests if user_interest == target_interest] 
```

```python
data_scientists_who_like("Java")
```

```python
from collections import defaultdict

# Keys are interests, values or lists of user_ids with that interest
user_ids_by_interest = defaultdict(list)

for user_id, interest in interests:
    user_ids_by_interest[interest].append(user_id)

print(user_ids_by_interest)
```
