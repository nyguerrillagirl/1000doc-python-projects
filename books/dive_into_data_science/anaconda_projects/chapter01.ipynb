## Analyzing Subsets of Data

```python
hour.loc[3, 'count']
```

```python
hour.loc[2:4,'registered']
```

```python
hour.loc[hour['hr'] < 5, 'registered'].mean()
```

```python
print(hour.loc[(hour['hr']<5) & (hour['temp']<.50),'count'].mean())
print(hour.loc[(hour['hr']<5) & (hour['temp']>.50),'count'].mean())
```

```python
# high temp or high humidity
print(hour.loc[(hour['temp']>0.5) | (hour['hum']>0.5),'count'].mean())
```
