pd.Series.idxmin() is one of those tiny-but-mighty pandas tools that shows up everywhere once you get comfortable with it. It returns the index label corresponding to the minimum value in a Series. Since you like example‑driven explanations, let’s walk through several patterns you’ll see in real data work.


```python
import numpy as np
import pandas as pd
```

## 1. Basic numeric Series

```python
import pandas as pd

s = pd.Series([10, 3, 7, 2, 9], index=["a", "b", "c", "d", "e"])
s
```

```python
s.idxmin()
```

## 2. When the index is meaningfule (e.g., product names)

```python
sales = pd.Series({
    "Keyboard": 120,
    "Mouse": 80,
    "Monitor": 200,
    "USB Cable": 40
})

sales
```

```python
sales.idxmin()
```

This pattern is common when you want to identify the worst‑performing item, lowest score, smallest count, etc.

## 3. With a DatetimeIndex

```python
import pandas as pd

temps = pd.Series(
    [32, 30, 28, 35, 40],
    index=pd.date_range("2024-01-01", periods=5)
)

temps
```

```python
temps.idxmin()
```

This is great for time‑series analysis: “On which day was the temperature lowest?”

## 4. With duplicate minimum values

If multiple entries share the minimum, pandas returns the first occurrence


```python
s = pd.Series([5, 1, 3, 1, 4], index=list("abcde"))
s.idxmin()
```

## 5. Using idxmin() after filtering

```python
scores = pd.Series([88, 92, 75, 91, 84], index=["Ann", "Ben", "Cara", "Dan", "Eli"])

# Find the lowest score among people whose name starts with a vowel
subset = scores[scores.index.str.startswith(("A", "E", "I", "O", "U"))]
subset
```

```python
subset.idxmin()
```

This pattern shows up constantly in data cleaning and exploratory analysis.

## 6. Using idxmin() inside a DataFrame operation

```python
df = pd.DataFrame({
    "Math": [90, 70, 85],
    "Science": [88, 95, 80],
    "English": [92, 78, 88]
}, index=["Alice", "Bob", "Cara"])

df
```

```python
df["Science"].idxmin()
```

You’ll use this a lot when selecting rows based on column minima.

## 7. idxmin() with missing values
By default, NaNs are ignores

```python
s = pd.Series([4, None, 2, None, 5], index=list("abcde"))
s
```

```python
s.idxmin()
```

If all values are NaN, pandas raises a ValueError.

```python
import pandas as pd
df = pd.DataFrame({'team': ['A', 'A', 'B', 'B'], 'position': ['GK', 'FW', 'GK', 'FW'], 'score': [1, 2, 3, 4]})
df
```

```python
pd.pivot_table(df, index='team', columns='position', values='score', aggfunc="max")
```
