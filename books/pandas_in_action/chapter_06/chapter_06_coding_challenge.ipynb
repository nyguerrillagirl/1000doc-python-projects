### 6.6.1 Problems

Our customers data set includes an Address column. Each address consists of a street,
a city, a state, and a zip code. Your challenge is to separate these four values; assign
them to new Street, City, State, and Zip columns; and then remove the Address column. 
Give the problem a shot, and then review the solution.

```python
import pandas as pd
```

```python
customers = pd.read_csv("customers.csv")
customers.head()
```

```python
customers["Address"].str.split(pat=", ", expand=True)
```

```python
customers[["Street", "City", "State", "Zip"]] = customers["Address"].str.split(pat=",", expand=True)
customers.head()
```

```python
customers = customers.drop(labels="Address", axis="columns")
customers.head()
```
