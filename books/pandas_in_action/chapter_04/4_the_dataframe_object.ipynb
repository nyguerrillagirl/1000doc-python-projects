```python
import pandas as pd
import numpy as np
```

### 4.1.1 Creating a DataFrame from a dictionary

```python
city_data = {
            "City": ["New York City", "Paris", "Barcelona", "Rome"],
            "Country": ["United States", "France", "Spain", "Italy"],
            "Population": [8600000, 2141000, 5515000, 2873000]
        }
cities = pd.DataFrame(city_data)
cities
```

```python
# What if we wanted to swap the column headers with the index labels
cities.transpose()
```

```python
cities.loc[0,:]
```

```python
type(cities.transpose())
```

```python
cities.transpose().loc['City', :]
```

```python
# This is the same as using transpose()
cities.T
```

### 4.1.2 Creating a DataFrame from a NumPy ndarray

```python
# create a 3 x 5 ndarray of integers between 1 and 101 (exclusive)
random_data = np.random.randint(1, 101, [3,5])
random_data
```

```python
pd.DataFrame(data=random_data)
```

```python
row_labels = ["Morning", "Afternoon", "Evening"]
temperatures = pd.DataFrame(data = random_data, index = row_labels)
temperatures
```

```python
column_labels = ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday")
pd.DataFrame(data= random_data, index=row_labels, columns = column_labels)
```

## 4.3 Sorting a DataFrame

### 4.3.1 Sorting by a single column

```python
nba.sort_values("Name")
```

```python
nba.sort_values("Name", ascending=False).head()
```

### 4.3.2 Sorting by multiple columns

```python
nba.sort_values(by=["Team", "Name"]).head()
```

```python
nba.sort_values(by=['Team', 'Salary'], ascending=[True, False])
```

```python
nba = nba.sort_values(by=['Team', 'Salary'], ascending=[True, False])
nba.head()
```
