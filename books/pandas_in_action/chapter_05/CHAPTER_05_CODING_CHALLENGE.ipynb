# Chapter 5 - Coding Challenge

The netflix.csv data set is a collection of almost 6,000 titles that were available to watch in November 2019 on the video streaming service Netflix.  
It includes four columns: 
 - the video’s title, 
 - director, 
 - the date Netflix added it, and 
- its type/category. 
    
The directorand date_added columns contain missing values. We can see examples at index positions 0, 2, and 5836 of the following output:

```python
import pandas as pd
df_netflix = pd.read_csv("netflix.csv", parse_dates=['date_added'])
```

```python
df_netflix.head()
```

```python
df_netflix.info()
```

```python
df_netflix['type'].value_counts()
```

## 1. Optimize the data set for limited memory use and maximum utility.

```python
df_netflix['type'] = df_netflix['type'].astype("category")
```

```python
df_netflix.info()
```

```python
df_netflix['director'].value_counts()
```

## 2. Find all rows with a title of "Limitless".

```python
df_netflix[df_netflix['title'] == 'Limitless']
```

## 3. Find all rows with a director of "Robert Rodriguez" and a type of "Movie".

```python
director_robert = df_netflix["director"] == "Robert Rodriguez"
movie_type = df_netflix["type"] == "Movie"
df_netflix[director_robert & movie_type]
```

## 4. Find all rows with either a date_added of "2019-07-31" or a director of "Robert Altman".

```python
date_2019_07_31 = df_netflix["date_added"] == "2019-07-31"
director_altman = df_netflix["director"] == "Robert Altman"
df_netflix[date_2019_07_31 | director_altman]
```

## 5. Find all rows with a director of "Orson Welles", "Aditya Kripalani", or "Sam Raimi"

```python
great_directors = ["Orson Welles", "Aditya Kripalani", "Sam Raimi"]
df_netflix[df_netflix["director"].isin(great_directors)]
```

## 6. Find all rows with a date_added value between May 1, 2019 and June 1, 2019

```python
df_netflix[df_netflix["date_added"].between(left="2019-05-01", right="2019-06-01")]
```

## 7. Drop all rows with a NaN value in the director column.

```python
df_netflix.dropna(subset=["director"])
```
