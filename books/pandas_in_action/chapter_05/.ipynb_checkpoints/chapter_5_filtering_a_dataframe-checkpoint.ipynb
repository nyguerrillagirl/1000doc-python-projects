## 5.3 Filtering my multiple conditions

### 5.3.1 The AND condition

```python
is_female = employees["Gender"] == "Female"
in_biz_dev = employees["Team"] == "Business Dev"
employees[is_female & in_biz_dev].head()
```

```python
is_manager = employees["Mgmt"]
employees[is_female & in_biz_dev & is_manager].head()
```

### 5.3.2 The OR condition

```python
earning_below_40k = employees["Salary"] < 40000
started_after_2015 = employees["Start Date"] > "2015-01-01"
employees[earning_below_40k | started_after_2015].tail()
```

### 5.3.3 Inversion with ~

```python
employees[employees["Salary"] < 100000].head()
```

```python
employees[~(employees["Salary"] >= 100000)].head()
```

### 5.3.4 Methods for Booleans

| Operation                | Arithmetic syntax                | Method Syntax                     |
|--------------------------|----------------------------------|-----------------------------------|
| Equality                 | employees["Team"] == "Marketing" | employees["Team"].eq("Marketing") |
| Inequality               | employees["Team"] != "Marketing" | employees["Team"].ne("Marketing") |
| Less than                | employees["Salary"] < 100000     | employees["Salary"].lt(100000)    |
| Less than or equal to    | employees["Salary"] <= 100000    | employees["Salary"].le(100000)    |
| Greater than             |  employees["Salary"] > 100000    | employees["Salary"].gt(100000)    |
| Greater than or equal to | employees["Salary"] >= 100000    | employees["Salary"].ge(100000)    |

## 5.4 Filtering by condition

### 5.4.1 The isin method

```python
sales = employees["Team"] == "Sales"
legal = employees["Team"] == "Legal"
mktg = employees["Team"] == "Marketing"
employees[sales | legal | mktg].head()
```

```python
all_star_teams = ["Sales", "Legal", "Marketing"]
on_all_star_teams = employees["Team"].isin(all_star_teams)
employees[on_all_star_teams].head()
```

### 5.4.2 The betweeen method

```python
# identify all employees with a salary between $80,000 and $90,000
higher_than_80 = employees["Salary"] >= 80000
lower_than_90 = employees["Salary"] < 90000
employees[higher_than_80 & lower_than_90].head()
```

```python
between_80k_and_90k = employees["Salary"].between(80000, 90000)
employees[between_80k_and_90k].head()
```

```python
# find all employees who started with the company in the 1980's
eighties_folk = employees["Start Date"].between(left="1980-01-01", right="1990-01-01")
employees[eighties_folk].head()
```

```python
# employees where first name starts with "R"
name_starts_with_r = employees["First Name"].between("R", "S")
employees[name_starts_with_r].head()
```

### 5.4.4 Dealing with null values

```python
# restore our dataset
employees = pd.read_csv("employees.csv", parse_dates=["Start Date"])
employees.head()
```

```python
employees.dropna()
```

```python
employees.tail()
```

```python
employees.dropna(how='all').tail()
```

```python
# use subnet parameter to target a column
employees.dropna(subset=["Gender"]).tail()
```

```python
# drop, specifying a list of columns
employees.dropna(subset=["Start Date", "Salary"]).head()
```

```python
# filter employees for rows with at least four present values
employees.dropna(thresh=4).head()
```

## 5.5 Dealing with duplicates

### 5.5.1 The duplicated method

```python
employees.head()
```

```python
employees["Team"].duplicated().head()
```

```python
employees["Team"].duplicated(keep="first").head()
```

```python
employees["Team"].duplicated(keep="last")
```

```python
(~employees["Team"].duplicated()).head()
```

```python
first_one_in_team = ~employees["Team"].duplicated()
employees[first_one_in_team]
```

### 5.5.2 The drop_duplicates method

```python
# find the first occurence of each unique value in the Team column
employees.drop_duplicates(subset = ["Team"])
```

```python
employees.drop_duplicates(subset = ["Team"], keep = "last")
```

```python
# filters for rows in employees with a unique value in the First Name column (first name appears only one time)
employees.drop_duplicates(subset = ["First Name"], keep = False)
```

```python
employees[employees["First Name"] == "Dennis"]
```

```python
# first occurence of each employee with a unique combination of "First Name" and "Gender" in the dataset
name_is_douglas = employees["First Name"] == "Douglas"
is_male = employees["Gender"] == "Male"
employees[name_is_douglas & is_male]
```

```python
employees.drop_duplicates(subset=["Gender", "Team"]).head()
```

```python
# find a male and female that share the same name
first_name_and_gender = employees.drop_duplicates(subset=["First Name", "Gender"])
first_name_and_gender.head()
```

```python
# number of employees with first name, gender
len(first_name_and_gender)
```
