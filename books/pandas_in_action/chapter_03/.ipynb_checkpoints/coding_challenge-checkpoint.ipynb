```python
import pandas as pd
```

Suppose that a historian reaches out to us and asks us to determine which day of the
week saw the most battles during the Revolutionary War. The final output should be a
Series with the days (Sunday, Monday, and so on) as index labels and a count of bat-
tles on each day as the values. Starting from scratch, import the revolutionary_war.csv
data set, and perform the necessary operations to arrive at the following data:

Saturday     39  
Friday       39  
Wednesday    32  
Thursday     31  
Sunday       31  
Tuesday      29  
Monday       27  

```python
df = pd.read_csv("revolutionary_war.csv", index_col='Start Date', parse_dates=['Start Date'], usecols=["State", "Start Date"])
battles = df.squeeze("columns")
battles.head()
```

```python
import datetime as dt
today = dt.datetime(2025, 7, 29)
today.strftime("%A")
```

```python
def day_of_week(adate):
    if adate is pd.NaT or pd.isna(adate):
        return 
    datetime_value = adate.to_pydatetime()
    return datetime_value.strftime("%A")
```

```python
day_of_week(battles.index[3])
```

```python
day_of_week(battles.index[4])
```

```python
battles.index.map(lambda x: day_of_week(x)).value_counts()
```
