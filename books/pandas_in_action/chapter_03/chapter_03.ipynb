```python
def single_or_multi(pokemon_type):
    if "/" in pokemon_type:
        return "Multi"
    return "Single"
```

```python
pokemon.head()
```

```python
pokemon.apply(single_or_multi)
```

```python
pokemon.apply(single_or_multi).value_counts()
```
