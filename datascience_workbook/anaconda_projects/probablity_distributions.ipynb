# Numpy probability distributions

A probability distribution describes the likelihood of different possible outcomes for a random variable, assigning a probability to each outcome such that the total probability sums to one. 

It serves as a mathematical model to describe the behavior of variables, helping to understand their characteristics and predict future events. 

Distributions are categorized as discrete for countable outcomes (like coin tosses) or continuous for outcomes that can take any value within a range (like an egg's weight). 

## Types of Probability Distributions
**Discrete Probability Distributions**:
These are used when the random variable can only take on specific, separate values. 

Probability Mass Function (PMF): This function maps each discrete outcome to its specific probability, for example, the probability of getting heads when flipping a coin.

**Continuous Probability Distributions:**

These are used for random variables that can take on any value within a given range. 

Probability Density Function (PDF): This function describes the probability of a continuous variable taking a specific value or falling within a certain range, where the area under the curve represents probability. 

### Examples
**Discrete:**
When you flip a coin twice, the probability distribution for the number of tails (0, 1, or 2) would show the likelihood of each specific outcome. 

**Continuous:**
An egg farmer might use a probability distribution to show the likelihood of different weights for eggs from their farm, where weights can vary by small increments. 

### Key Characteristics
- The probability of any specific outcome must be between 0 and 1. 
- The sum of probabilities for all possible outcomes must equal 1. 
- Probability distributions are theoretical ideals of frequency distributions, used to model real-world phenomena and populations. 

## 1. Uniform Distribution

A good example to keep in mind is the possible outcomes when flipping a coin (heads or tails) or a die (1, 2, 3, 4, 5, 6)

```python
# simulate throwing a die
import numpy as np
roll = int(np.random.uniform(1,7)) # 7 is exclusive
print(roll)
```

```python
# roll again
roll = int(np.random.uniform(1,7))
print(roll)
```

```python
# multiple rolls
rolls = [int(np.random.uniform(1,7)) for _ in range(10)]
print(rolls)
```

```python
# better alternative
roll = np.random.randint(1, 7)  # Discrete uniform distribution
print(roll)
```

```python
# another way
roll = np.random.choice([1, 2, 3, 4, 5, 6])
print(roll)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# generate many rolls from a uniform distribution and create a count plot to see if it "looks" uniform
# let's try 1000
rolls = np.random.randint(1, 7, size=1000)
sns.countplot(x=rolls)
plt.title("Uniform Distribution of 1000 Die Rolls")
plt.xlabel("Die Face")
plt.ylabel("Frequency")
plt.show()
```

```python
# let's run it 100,000 times to see if it appears more uniform
rolls = np.random.randint(1, 7, size=100000)
sns.countplot(x=rolls)
plt.title("Uniform Distribution of 1000 Die Rolls")
plt.xlabel("Die Face")
plt.ylabel("Frequency")
plt.show()
```

## 2. Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
mu = 0       # Mean
sigma = 1    # Standard deviation
n = 1000     # Number of samples

# Generate normally distributed data
data = np.random.normal(loc=mu, scale=sigma, size=n)

# Visualize with a histogram and KDE
sns.histplot(data, bins=30, kde=True, color="skyblue")
plt.title("Normal Distribution (μ=0, σ=1)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```
