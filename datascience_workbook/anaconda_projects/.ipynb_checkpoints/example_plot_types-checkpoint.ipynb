```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
tips = sns.load_dataset("tips")

# Line plot of total bill over index
sns.lineplot(data=tips, x=tips.index, y="total_bill")
plt.title("Line Plot of Total Bill")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = sns.load_dataset("iris")
```

```python
iris.info()
```

```python
# Drop non-numeric column
iris_numeric = iris.drop(columns=['species'])

# Compute correlation matrix
corr = iris_numeric.corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Iris Dataset)")
plt.show()
```

```python
# example of scatterplot
# Load dataset
tips = sns.load_dataset("tips")

# Scatter plot of total bill vs tip
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
# example linear regression plot

# Load dataset
tips = sns.load_dataset("tips")

# Linear regression plot
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("Linear Regression of Tip vs Total Bill")
plt.show()
```
